```python
import os
import numpy as np
from skimage import io
%matplotlib
```

```python
data_path = './hw6_data/Aberdeen/'
```

```python
imgs_path = os.listdir(data_path)
imgs = []
for path in imgs_path:
    imgs.append((io.imread(data_path + path)).flatten())

imgs = np.array(imgs)
x_mean = imgs.mean(axis=0)
img_mean = x_mean.astype(np.uint8).reshape(600, 600, 3)
io.imshow(img_mean)
```

```python
X = (imgs - x_mean).T
U, s, V = np.linalg.svd(X, full_matrices=False)
```

```python
e = U.T[9]
e = e * -1
e -= np.min(e)
e /= np.max(e)
e = (e * 255).astype(np.uint8)
e = e.reshape(600, 600, 3)
io.imshow(e)
```
