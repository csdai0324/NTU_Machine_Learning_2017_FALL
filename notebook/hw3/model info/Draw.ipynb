```python
import matplotlib.pyplot as plt
```

```python
def read_history(history_path):
    loss = []
    val_loss = []
    acc = []
    val_acc = []
    with open(history_path, 'r') as his:
        linecount = 0
        for line in his:
            if linecount % 2 == 0:
                linecount += 1
                continue
            lt = line.strip().split(' ')
            loss.append(float(lt[7]))
            acc.append(float(lt[10]))
            val_loss.append(float(lt[13]))
            val_acc.append(float(lt[16]))
            linecount += 1
    return loss, val_loss, acc, val_acc
```

```python
loss, val_loss, acc, val_acc = read_history('./dnn history')
```

```python
fig1 = plt.figure(1, dpi=100)
epoch = list(range(0, 250))
plt.plot(epoch, loss, label='loss')
plt.plot(epoch, val_loss, label='val_loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()

fig2 = plt.figure(2, dpi=100)
plt.plot(epoch, acc, label='acc')
plt.plot(epoch, val_acc, label='val_acc')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend()
plt.show()

fig1.savefig('dnn_loss.png')
fig2.savefig('dnn_acc.png')
```
